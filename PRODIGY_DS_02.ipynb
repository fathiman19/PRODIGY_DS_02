```python
#importing necessory libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data
df = pd.read_csv('/content/test (1).csv')
```

```python
df.dropna(subset=["Embarked"],inplace = True)
df["Cabin"].fillna("Unknown",inplace = True)
df["Age"].fillna(df["Age"].mean(),inplace = True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Age"],kde = True, color = 'purple')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure (figsize = (10,6))
sns.countplot(x = 'Sex',data = df, hue ="Sex", palette = "Set2")
plt.title("Survival by gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title = "Survived", loc ="upper right")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(data = df, x = "Age", y ="Fare", hue = "Age", palette = "viridis")
plt.title("Scatter plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title = "Age", loc = "upper right")
plt.show()
```
